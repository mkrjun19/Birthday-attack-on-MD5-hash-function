```python
import hashlib
import string    
import random
import datetime
import time
```

```python
sum=0;
for i in range(10):
  st=datetime.datetime.now()
  d={}
  while(1):
      S = 20  
      ran = ''.join(random.choices(string.ascii_uppercase + string.ascii_lowercase + string.digits, k = S))    
      result = hashlib.md5(str(ran).encode())
      result=result.hexdigest()[:8]
      if result not in d.values():
        d[str(ran)]=result
      else:
          print(str(ran))
          for key, value in d.items():
            if result == value:
              print(key)
          print("the above two strings have same first 32bits hash value :",result)
          sum=sum+len(d)
          et=datetime.datetime.now()
          it=et-st
          print("Found in",len(d),"iterations and time taken is",it,"(Hours:mins:seconds)")
          print("\n")
          break
print("so we needed average of",sum/10,"Iterations to break MD5")
```
